```python
from gensim.models import word2vec
```

```python
text = []
with open('/home/shotaro/Dropbox/log.txt') as f:
    for s in f.readlines():
        text += s.split()[1:]
with open('/home/shotaro/copus.txt', 'w') as w:
    for t in text:
        w.write(t)
        w.write(" ")
```

```python
data = word2vec.Text8Corpus('/home/shotaro/copus.txt')
```

```python
model = word2vec.Word2Vec(data, size=4, min_count=1)
```
